```python
df = df.drop(['Price Starting With ($)','Publish Date (Month)','Publish Date (Year)'], axis=1)
```

```python

def decapi(x):
    if not isinstance(x, str):
        return x
    

    x = x.lower()
    

    for i in x:
        if not i.isalnum() :
            x = x.replace(i, "")
    
    # Remove '&' and 'by'
    x = x.replace("&", "")
    x = x.replace("by", "")
    
    return x
```

```python

columns_to_clean = ['Authors', 'Publisher']
for col in columns_to_clean:
    if col in df.columns:
        df[col] = df[col].apply(decapi)
```

```python
df
```

```python
y = ['Description', 'Category']
for col in y:
    if col in df.columns:
        df[col] = df[col].str.lower().str.replace(r'[-"â€™,]', ' ', regex=True).str.replace(r"'s\b", '', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip() 
```

```python
df
```

```python
def tolist(h):
    if isinstance(h, list):
        return h
    elif isinstance(h, str):
        return h.split()
    else:
        return [h] 
    
```

```python
z = ['Authors', 'Publisher','Description','Category','Publisher']
for col in z:
    if col in df.columns:
        df[col] = df[col].apply(tolist)
```

```python
df['tags']=df['Authors']+df['Description']+df['Category']+df['Publisher']
```

```python
new_df=df[['Title','tags']]
```

```python
new_df.loc[:, 'id'] = range(0, len(new_df))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x : " ".join(x))
```

```python
import sys
# Use sys.executable to ensure pip installs into the correct environment
!{sys.executable} -m pip install nltk

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return "".join(y)
     
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
!pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vc = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vector=vc.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vc.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi=cosine_similarity(vector)
```

```python
def recon(book):
    book_index=new_df[new_df['Title']==book].index[0]
    dis=simi[book_index]
    book_list=sorted(list(enumerate(dis)),reverse=True,key=lambda x : x[1])[1:11]
    for i in book_list:
        print(new_df.iloc[i[0]].Title)
```

```python
#sample input

recon('The New Face of Terrorism: Threats from Weapons of Mass Destruction')
```
