```python
import pandas as pd
from langchain.schema import Document
```

```python
df = pd.read_csv("BooksDatasetClean.csv")
```

```python
df
```

```python
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df = df.drop(['Price Starting With ($)','Publish Date (Month)','Publish Date (Year)'], axis=1)
```

```python
df.columns
```

```python
records = df.to_dict("records")
```

```python
documents = []
for record in records:
    # 1. Augment and define the Content (page_content)
    # Combining Title, Authors, and Description ensures the embedding is rich.
    page_content = (
        f"Title: {record['Title']}. "
        f"Authors: {record['Authors']}. "
        f"Description: {record['Description']}"
        f"category': {record['Category']}"
    )

    # 2. Define the Metadata (for filtering, not for semantic similarity)
    metadata = {
        'category': record['Category'],
        'publisher': record['Publisher'],
        'source_title': record['Title'] # Often useful to keep the title here too
    }
    doc = Document(page_content=page_content, metadata=metadata)
    documents.append(doc)
```
